```python
import os
import json
import nbformat
from datetime import datetime
from pathlib import Path

# å¯¾è±¡ãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒª
target_dir = Path("./02_notebook")
target_dir.mkdir(exist_ok=True)

# ä»Šæ—¥ã®æ—¥ä»˜
today = datetime.now().strftime("%Y%m%d")
new_filename = f"my_notebook_{today}.ipynb"
new_filepath = target_dir / new_filename

# æ–°ã—ã„ãƒãƒ¼ãƒˆãƒ–ãƒƒã‚¯ã‚’ä½œæˆ
notebook_content = {
    "cells": [],
    "metadata": {},
    "nbformat": 4,
    "nbformat_minor": 5
}
with open(new_filepath, "w", encoding="utf-8") as f:
    json.dump(notebook_content, f, indent=2)

print(f"{new_filename} ã‚’ä½œæˆã—ã¾ã—ãŸï¼")

# ä¸€è¦§ãƒãƒ¼ãƒˆãƒ–ãƒƒã‚¯ã‚’æ›´æ–°
notebooks = [
    f for f in os.listdir(target_dir)
    if f.endswith(".ipynb") and f != new_filename and f != "notebook_index.ipynb"
]
links_md = "\n".join([f"- [{nb}](./{nb})" for nb in sorted(notebooks)])

index_nb = nbformat.v4.new_notebook()
index_nb.cells.append(nbformat.v4.new_markdown_cell("# ğŸ“š ä»–ã®ãƒãƒ¼ãƒˆãƒ–ãƒƒã‚¯ä¸€è¦§\n\n" + links_md))

index_path = target_dir / "notebook_index.ipynb"
with open(index_path, "w", encoding="utf-8") as f:
    nbformat.write(index_nb, f)

print("notebook_index.ipynb ã‚’æ›´æ–°ã—ã¾ã—ãŸï¼")
```
