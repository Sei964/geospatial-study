```python
import json
from datetime import datetime

# 日付付きファイル名
today = datetime.now().strftime("%Y%m%d")
filename = f"my_notebook_{today}.ipynb"

# 最小限のJupyterノートブック構造
notebook_content = {
    "cells": [],
    "metadata": {},
    "nbformat": 4,
    "nbformat_minor": 5
}

# JSONとして保存
with open(filename, "w", encoding="utf-8") as f:
    json.dump(notebook_content, f, indent=2)

print(f"{filename} を作成しました！")
```
